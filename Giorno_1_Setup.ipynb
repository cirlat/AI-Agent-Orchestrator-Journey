```python
# 1. Definiamo i dati (Immagina che arrivino da un vecchio DB Windows)
record_legacy = "001ROSSI      MARIO     MILANO    20240520"

# 2. Parsing moderno (Slicing) - Estraiamo i dati con la logica posizionale che conosci
id_utente = record_legacy[0:3]
cognome = record_legacy[3:13].strip() # .strip() pulisce gli spazi vuoti, come una TRIM
nome = record_legacy[13:23].strip()

# 3. Trasformiamo in una lista per l'IA
dati_per_agente = [id_utente, cognome, nome]

# 4. Stampiamo il risultato
print(f"Dati estratti e pronti per l'IA: {dati_per_agente}")

# SFIDA: Aggiungi una logica che verifichi se il cognome è "ROSSI"
# e stampa "Utente trovato"
if cognome == "ROSSI":
    print("Utente trovato!")
```
