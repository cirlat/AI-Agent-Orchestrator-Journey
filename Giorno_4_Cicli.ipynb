```python
# Lista di log del server (simuliamo un dump di file TXT)
error_count=0
logs = [
    {"timestamp": "09:00", "status": "OK", "msg": "Servizio avviato"},
    {"timestamp": "09:05", "status": "ERROR", "msg": "Database timeout"},
    {"timestamp": "09:10", "status": "WARN", "msg": "Memoria quasi piena"},
    {"timestamp": "09:15", "status": "ERROR", "msg": "Accesso negato"},
]

print("--- ANALISI LOG PER AGENTE IA ---")

# Ciclo FOR per scansionare ogni singolo log
for log in logs:
    # Logica decisionale (IF)
    if log["status"] == "ERROR":
        # Se c'è un errore, prepariamo il "Task" per l'agente
        print(f"ALERTA IA: Analizzare l'errore '{log['msg']}' delle ore {log['timestamp']}")
        error_count += 1
    elif log["status"] == "WARN":
        print(f"NOTA: Monitorare {log['msg']}")
    else:
        print(f"SISTEMA OK: {log['timestamp']}")

# SFIDA: Prova a modificare il codice per contare quanti errori ci sono nella lista

print(f"Ci sono {error_count} errori")
```
