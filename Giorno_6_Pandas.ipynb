```python
import pandas as pd
import datetime

# 1. Creiamo un "Archivio Clienti" (Simuliamo un dump da un DB Windows)
data = {
    'ID_Cliente': [101, 102, 103, 104],
    'Nome': ['Mario Rossi', 'Luigi Bianchi', 'Elena Verdi', 'Sconosciuto'],
    'Acquisti_EUR': [1200, 450, 3200, 0],
    'Ultimo_Accesso': ['2023-01-10', '2023-05-12', '2024-02-01', None]
}

df = pd.DataFrame(data)

df['Ultimo_Accesso']=df['Ultimo_Accesso'].fillna("Mai")
df.loc[3,'Nome']= "Franco"
df.loc[3,'Acquisti_EUR']= 5555
df.loc[3,'Ultimo_Accesso']=datetime.date.today()

# 2. Visualizzazione (Molto più elegante di una DISPLAY COBOL)
print("--- PRIME RIGHE DEL DATAFRAME ---")
print(df.head())

# 3. Analisi istantanea (Il sogno di ogni manager)
print("\n--- STATISTICHE RAPIDE ---")
print(df.describe()) # Ti dà media, max, min degli acquisti in un colpo solo

# 4. Filtro intelligente: Chi sono i "Top Spenders" per l'IA?
top_customers = df[df['Acquisti_EUR'] > 1000]
print("\n--- CLIENTI TARGET PER L'AGENTE IA ---")
print(top_customers)
```
