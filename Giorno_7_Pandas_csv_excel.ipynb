```python
import pandas as pd

# 1. Carichiamo un file di esempio direttamente dal web (Dati sulle spedizioni)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df.describe())


df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Cabin'] = df['Cabin'].fillna("Clandestino")

print(f"età media dei passeggeri: {df['Age'].mean()}")

print("Passeggeri per classe:")

print(df['Cabin'].value_counts())


# 2. Esplorazione rapida (Cosa abbiamo caricato?)
print(f"Il file ha {df.shape[0]} righe e {df.shape[1]} colonne.")

# 3. Analisi: Quanti passeggeri ci sono per ogni classe?
# È come una rottura di livello con conteggio
print("\nDistribuzione Classi:")
print(df['Pclass'].value_counts())

# 4. Esportazione: Immagina di aver pulito i dati e di volerli salvare per l'Agente IA
# Lo salviamo in formato Excel (necessita di una libreria che Colab ha già)
df.to_excel("dati_puliti_per_ia.xlsx", index=False)
print("\nFile Excel creato con successo!")
```
