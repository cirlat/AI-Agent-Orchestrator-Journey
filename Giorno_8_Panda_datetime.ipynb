```python
from typing_extensions import Text
import pandas as pd

# 1. Dati grezzi con date in formato testo (come uscirebbero da un log)
data = {
    'Evento': ['Login', 'Update', 'Error', 'Login', 'Backup'],
    'Data_Ora': ['2024-05-01 08:30', '2024-05-02 12:00', '2024-05-02 15:45', '2024-05-03 09:00', '2024-05-04 23:00']
}
df = pd.DataFrame(data)

# 2. Conversione: Da testo a Datetime (Fondamentale!)
df['Data_Ora'] = pd.to_datetime(df['Data_Ora'])

# 3. Estrazione: L'IA può ora capire i singoli componenti
df['Giorno'] = df['Data_Ora'].dt.day_name() # Es: 'Monday'
df['Ora'] = df['Data_Ora'].dt.hour

# 4. Filtro Temporale: Vediamo solo gli eventi dopo il 2 Maggio
filtro_data = df[df['Data_Ora'] > '2024-05-02']
differenza_totale = df['Data_Ora'].max() - df['Data_Ora'].min()
print(f"giorni di differenza tra il e l\'ultimo giorno: {differenza_totale.days}")

print("--- ANALISI TEMPORALE ---")
print(df)
print("\n--- EVENTI DOPO IL 2 MAGGIO ---")
print(filtro_data)
# SFIDA: Calcola quanti giorni passano tra il primo e l'ultimo evento della tabella
# Suggerimento: usa df['Data_Ora'].max() - df['Data_Ora'].min()
```
